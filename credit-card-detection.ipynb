```python
import datetime
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.model_selection import train_test_split 
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score 
from sklearn.preprocessing import LabelEncoder
```

# Read Data

```python
td = pd.read_csv("/kaggle/input/fraud-detection/fraudTrain.csv")
```

# Preprocessing

```python
td["trans_date_trans_time"] = pd.to_datetime(td["trans_date_trans_time"])
td["dob"] = pd.to_datetime(td["dob"])
td
```

```python
td.drop(columns=['Unnamed: 0','cc_num','first', 'last', 'street', 'city', 'state', 'zip', 'dob', 'trans_num','trans_date_trans_time'],inplace=True)
td
```

```python
td.dropna(ignore_index=True)
td
```

```python
td.dropna(ignore_index=True)
```

```python
encoder = LabelEncoder()
td["merchant"] = encoder.fit_transform(td["merchant"])
td["category"] = encoder.fit_transform(td["category"])
td["gender"] = encoder.fit_transform(td["gender"])
td["job"] = encoder.fit_transform(td["job"])
```

```python
td
```

# EDA

```python
exit_counts = td["is_fraud"].value_counts()
plt.figure(figsize=(12, 8))
plt.subplot(1, 2, 1)  # Subplot for the pie chart
plt.pie(exit_counts, labels=["No", "YES"], autopct="%0.0f%%")
plt.title("is_fraud Counts")
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

# Train the model

```python
X = td.drop(columns=["is_fraud"], inplace = False)
Y = td["is_fraud"]
```

```python
model = SVC()
model.fit(X, Y)
```

```python
model.score(X, Y)
```

# Test the model

```python
tsd = pd.read_csv("/kaggle/input/fraud-detection/fraudTest.csv")
tsd
```

```python
tsd.drop(columns=['Unnamed: 0','cc_num','first', 'last', 'street', 'city', 'state', 'zip', 'dob', 'trans_num','trans_date_trans_time'],inplace=True)
tsd
```

```python
encoder = LabelEncoder()
tsd["merchant"] = encoder.fit_transform(tsd["merchant"])
tsd["category"] = encoder.fit_transform(tsd["category"])
tsd["gender"] = encoder.fit_transform(tsd["gender"])
tsd["job"] = encoder.fit_transform(tsd["job"])
```

```python
tsd
```

```python
X_test = tsd.drop(columns=["is_fraud"], inplace = False)
Y_test = tsd["is_fraud"]
```
